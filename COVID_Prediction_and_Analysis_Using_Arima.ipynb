```
#Prediction
#Creating rolling mean and stadard diviation over 7 days (week) : Pineham
rollmean = dfCovid19DatesPIIdx.rolling(window=7).mean()
rollstd = dfCovid19DatesPIIdx.rolling(window=7).std()
print(rollmean, rollstd)

#Creating rolling mean and stadard diviation over 7 days (week) : Basingstoke
rollmeanBA = dfCovid19DatesBAIdx.rolling(window=7).mean()
rollstdBA = dfCovid19DatesBAIdx.rolling(window=7).std()
print(rollmean, rollstd)

#Creating rolling mean and stadard diviation over 7 days (week) : Haydock
rollmeanHD = dfCovid19DatesHDIdx.rolling(window=7).mean()
rollstdHD = dfCovid19DatesHDIdx.rolling(window=7).std()
print(rollmean, rollstd)

#Creating rolling mean and stadard diviation over 7 days (week) : Hams Hall
rollmeanHH = dfCovid19DatesHHIdx.rolling(window=7).mean()
rollstdHH = dfCovid19DatesHHIdx.rolling(window=7).std()
print(rollmean, rollstd)

#Creating rolling mean and stadard diviation over 7 days (week) : Rye Park
rollmeanRP = dfCovid19DatesRPIdx.rolling(window=7).mean()
rollstdRP = dfCovid19DatesRPIdx.rolling(window=7).std()
print(rollmean, rollstd)

#Creating rolling mean and stadard diviation over 7 days (week) : Waltham Point
rollmeanWP = dfCovid19DatesWPIdx.rolling(window=7).mean()
rollstdWP = dfCovid19DatesWPIdx.rolling(window=7).std()
print(rollmean, rollstd)
```

```
#Function for calculating dickey fuller and plotting the results
def test_stationary(timeseries):
  #determining rolling statistics
  MovingAv = timeseries.rolling(window=7).mean()
  movingSTD = timeseries.rolling(window= 7).std()

  #plot the statistics
  orig = plt.plot(timeseries, color='blue', label='Original')
  mean = plt.plot(MovingAv, color='black', label='Rolling Mean')
  std  = plt.plot(MovingStd, color='red', label='Rolling STD')
  plt.legend(loc='best')
  plt.title('Rolling Mean & Standard Deviation')
  plt.xticks(rotation=45)
  plt.show(block=False)

  #Peforming Dickey-Fuller test again:
  print('Results of Dickey-Fuller Test:')
  dftest = adfuller(timeseries, autolag='AIC')
  dfoutput = pd.Series(dftest[0:4], index=['Test Stats', 'P-Val', 'Lags', 'Number of Observations Used'])
  for key, valye in dftest[4].items():
    dfoutput['Critical Value (%s)' %key] = value
  print(dfoutput)
```

```
#Testing the transformed data with dickey fuller test
test_stationary(dfCovid19DatesPI_logScale)
test_stationary(dfCovid19DatesBA_logScale)
test_stationary(dfCovid19DatesHD_logScale)
test_stationary(dfCovid19DatesHH_logScale)
test_stationary(dfCovid19DatesRP_logScale)
test_stationary(dfCovid19DatesWP_logScale)
```
